```python
#look at using each station number
list_of_stations = df['Station Number'].unique()
#list_of_stations.sort() - can sort them (1-102)
print(list_of_stations)
```

```python
####COOL CODE#####
##this shows the flucuation in bikes over the 40 min period

for i in list_of_stations:
    get_stat = df.loc[df['Station Number'] == i]
    title = get_stat['Name'].unique()
    get_stat = get_stat.drop(['Number','Address','Banking','Bike Stands','Bonus','Contract Name','First Update','Status','Last Update','Position','Name','Station Number'], axis=1)

    #####comment out for now
    get_stat.plot.bar(title=title)
    plt.savefig("Bikes_Stands_Avail" + str(i) + ".png")
```
