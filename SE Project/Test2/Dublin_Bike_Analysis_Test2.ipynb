```python
#change station number to categorical
test['Station Number'] = test['Station Number'].astype('object')
test.dtypes
```

```python
dates = []
times = []

for line in test['DateTime']:
    line = str(line)
    n = 8
    split_line = [line[i:i+n] for i in range(0, len(line), n)]
    #print(split_line)
    date = split_line[0]
    time = split_line[1]
    print(time[0],time[1], time[2])
    time = str(time[0]) + str(time[1]) + str(time[2])
    print(time)
    
    dates.append(date)
    times.append(time)
    
test["Date"] = dates
test["Time"] = times
```

```python
#now remove old datetime column

test = test.drop(["DateTime"], axis = 1)
```

```python
unique_times = test['Time'].unique()
```

```python
#find station with the most bikes at a specific time
%matplotlib inline

#first for loop makes new df based on each time
for time in unique_times:
    time_frame = test.loc[test['Time'] == time]
    #now analyse each time frame
    A = time_frame['Available Bike Stands']
    B = time_frame['Bike Stands']
    percentage_available_stands = (A/B) * 100
    
    #find highest % avail bikes
    ###print("% available bike stands at time: ", time, "is: ", percentage_available_stands)
    high_bikes = max(percentage_available_stands)
    
    #add this col to time frame
    test['% Available Stands'] = percentage_available_stands
    
    #check names with highest % available bikes in each time frame
    empty_stands = time_frame.loc[test['% Available Stands'] == high_bikes]
    #print(empty_stands['Name'], empty_stands['Station Number'], high_bikes) # every 5 mins so alot of the stations dont change
    cleaned_empty_stands = empty_stands.drop(['Number','Address','Available Bike Stands','Available Bikes','Banking','Bike Stands','Bonus','Contract Name','First Update',
                                              'Station Number','Status','Position','Date','Time'], axis=1)
    #cleaned_empty_stands.plot.bar(figsize=(20, 20))
    empty_stands.plot.bar(figsize=(20,20))
```
